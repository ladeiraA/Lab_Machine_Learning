```python
# Importar bibliotecas
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
# Carregar dados
df = pd.read_csv("filmes.csv", index_col=0)
df.head()
```

```python
# Remover colunas 
df = df.drop("Label", axis=1)
df = df.drop("Movie Name", axis=1)
df.head()
```

```python
# Preparar o algoritimo
knn = NearestNeighbors(n_neighbors=3)
knn.fit(df)
```

```python
knn.kneighbors(([]))
```
